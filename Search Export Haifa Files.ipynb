```python
from dataflows import Flow, load, checkpoint, printer

Flow(
    load('data/search_export_haifa_files/datapackage.json'),
    printer(tablefmt='html', num_rows=1, resources=['search_export_haifa_files'], fields=['title', 'bib_title'])
).process()[1]
```

```python
from tabulator import Stream
import json, yaml
from nli_z3950.parse_bib_data import parse_marc_bib_data

def print_yaml(data):
    print(yaml.dump(data, allow_unicode=True, default_flow_style=False))
    
with Stream('data/search_export_haifa_files/search_haifa_files.csv', headers=1) as stream:
    for i, row in enumerate(stream.iter(keyed=True)):
        # if i > 10: break
        m = json.loads(row['json'])
        bib_data = parse_marc_bib_data(m)
        if i % 1000 == 0: print_yaml(bib_data)
```
